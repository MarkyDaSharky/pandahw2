```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('./AB_NYC_2019.csv')
air_bnb
```

```python
# How many neighbourhood groups / which shows up the most
air_bnb.groupby('neighbourhood_group').count().sort_values(['id'], ascending=False)[['id']]
"""using count(), totaling all id's Manhattan showed up the most"""
```

```python
# Are private rooms the most popular in Manhattan?
air_bnb.groupby('neighbourhood_group').count()
```

```python
# Who are the top 10 busiest hosts based on number of reviews per month?
top_hosts = air_bnb.groupby(['host_name']).sum()
top10 = top_hosts.sort_values('reviews_per_month', ascending=False)[['reviews_per_month']].head(10)
top10
```

```python
# Which neighborhood group has the highest avg price and plot the average prices in a bar chart
air_bnb.groupby('neighbourhood_group').mean()[['price']].sort_values(['price'], ascending=False)
#Manhattan has the highest price 

```

```python
# What is the most expensive neighbourhood in each neighbourhood_group
air_bnb.groupby('neighbourhood').mean().sort_values('price', ascending=False)[['price']]
```

```python
# Which host has the highest minimum rental? *minimal rental is based on price and minimum nights
```
